```python
Oklahoma, = plt.plot(x_axis, Ok, marker="+",color="blue", linewidth=1, label="Oklahoma")
Flordia, = plt.plot(x_axis, Fl, marker="s", color="Red", linewidth=1, label="Flordia")
Kentucky, = plt.plot(x_axis, Ky, marker="o",color="indigo", linewidth=1, label="Kentucky")
Missouri, = plt.plot(x_axis, Mo, marker="d", color="darkorange", linewidth=1, label="Missouri")
Texas, = plt.plot(x_axis, Tx, marker="<",color="darkgreen", linewidth=1, label="Texas")
```
