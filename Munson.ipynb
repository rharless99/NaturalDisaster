```python
import requests
import json
import pandas as pd
from datetime import datetime,date
```

```python
base_url = "https://eonet.sci.gsfc.nasa.gov/api/v2.1/events?days = 5000"
response = requests.get(base_url)
Nasa_data = response.json()
print(json.dumps(Nasa_data, indent=4, sort_keys=True))
```

```python
titles=[]
lats =[]
lngs = []
date = []
for num in range(len(Nasa_data["events"])):
    titles.append(Nasa_data["events"][num]["categories"][0]['title'])
    lats.append(Nasa_data["events"][num]["geometries"][0]["coordinates"][0])
    lngs.append(Nasa_data["events"][num]["geometries"][0]["coordinates"][-1])
    date.append(Nasa_data["events"][num]["geometries"][0]['date'])
date
```

```python
disasters_df = pd.DataFrame({"Type": titles,
                            "Latitude" : lats,
                            "Longitude" : lngs,
                            "Date" : date})
disasters_df
```

```python
disasters_df['Date'] = disasters_df['Date'].str.strip()
```
