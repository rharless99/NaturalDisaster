```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
weather_file = "Weather_CSV/Weather.csv"
disaster_file = "disaster_count_by_year&state.csv"
```

```python
Weather_df = pd.read_csv(weather_file)
Weather_df.head()
```

```python
Weather_df = Weather_df.drop(columns=['Average Temperature', 'Cloud Cover', 'Relative Humidity','Wind Speed', 'Precipitation'])
```

```python
Weather_df_reduced = Weather_df.dropna(how='any')
```

```python
Weather_df_reduced = Weather_df_reduced.drop(columns=['Month'])
```

```python
Weather_df_years = Weather_df_reduced.loc[Weather_df_reduced["Year"]<=2010]
```

```python
Weather_df_years
```

```python
disasters_df = pd.read_csv(disaster_file)
disasters_df.head()
```

```python
x_axis = disasters_df['Year']
```

```python
OK = disasters_df['OK']
FL = disasters_df['FL']
MO = disasters_df['MO']
KY = disasters_df["KY"]
TX = disasters_df['TX']
Ok = Weather_df_years.loc[Weather_df_years["Location"]=="Oklahoma"]["Yearly Average Temp"]
Fl = Weather_df_years.loc[Weather_df_years["Location"]=="Florida"]["Yearly Average Temp"]
Ky = Weather_df_years.loc[Weather_df_years["Location"]=="Kentucky"]["Yearly Average Temp"]
Mo = Weather_df_years.loc[Weather_df_years["Location"]=="Missouri"]["Yearly Average Temp"]
Tx = Weather_df_years.loc[Weather_df_years["Location"]=="Texas"]["Yearly Average Temp"]
OkRain = Weather_df_years.loc[Weather_df_years["Location"]=="Oklahoma"]["Average Precipitation"]
FlRain = Weather_df_years.loc[Weather_df_years["Location"]=="Florida"]["Average Precipitation"]
KyRain = Weather_df_years.loc[Weather_df_years["Location"]=="Kentucky"]["Average Precipitation"]
MoRain = Weather_df_years.loc[Weather_df_years["Location"]=="Missouri"]["Average Precipitation"]
TxRain = Weather_df_years.loc[Weather_df_years["Location"]=="Texas"]["Average Precipitation"]
```

```python
#OKLAHOMA
plt.plot(x_axis, Ok, marker="+",color="blue", linewidth=1, label="OK Temperature")
plt.legend(loc="best")
plt.xlabel("Years")
plt.ylabel("Average Yearly Temperature (F)")
plt.title("Average Yearly Temperature Changes")
plt.savefig("Images/OK_YearlyTemp_Change.png")
plt.xticks(Rotation=45)
plt.show
```
